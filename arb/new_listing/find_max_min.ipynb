```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv("xaiusdt_price_comparison (1).csv")

# Find the max and min in the 'Percentage Difference' column
max_difference = df['Percentage Difference'].max()
min_difference = df['Percentage Difference'].min()

# Print max and min differences
print(f"Max Percentage Difference: {max_difference}")
print(f"Min Percentage Difference: {min_difference}")

# Plot the 'Percentage Difference'
plt.figure(figsize=(12, 6))
plt.plot(df['Percentage Difference'], label='Percentage Difference')
plt.axhline(y=max_difference, color='r', linestyle='-', label=f'Max Difference ({max_difference})')
plt.axhline(y=min_difference, color='g', linestyle='-', label=f'Min Difference ({min_difference})')
plt.xlabel('Index')
plt.ylabel('Percentage Difference')
plt.title('Percentage Difference over Time')
plt.legend()
plt.show()





# Load the data
df = pd.read_csv("aiusdt_price_comparison (1).csv")

# Find the max and min in the 'Percentage Difference' column
max_difference = df['Percentage Difference'].max()
min_difference = df['Percentage Difference'].min()

# Print max and min differences
print(f"Max Percentage Difference: {max_difference}")
print(f"Min Percentage Difference: {min_difference}")

# Plot the 'Percentage Difference'
plt.figure(figsize=(12, 6))
plt.plot(df['Percentage Difference'], label='Percentage Difference')
plt.axhline(y=max_difference, color='r', linestyle='-', label=f'Max Difference ({max_difference})')
plt.axhline(y=min_difference, color='g', linestyle='-', label=f'Min Difference ({min_difference})')
plt.xlabel('Index')
plt.ylabel('Percentage Difference')
plt.title('Percentage Difference over Time')
plt.legend()
plt.show()





# Load the data
df = pd.read_csv("nfpusdt_price_comparison (2).csv")

# Find the max and min in the 'Percentage Difference' column
max_difference = df['Percentage Difference'].max()
min_difference = df['Percentage Difference'].min()

# Print max and min differences
print(f"Max Percentage Difference: {max_difference}")
print(f"Min Percentage Difference: {min_difference}")

# Plot the 'Percentage Difference'
plt.figure(figsize=(12, 6))
plt.plot(df['Percentage Difference'], label='Percentage Difference')
plt.axhline(y=max_difference, color='r', linestyle='-', label=f'Max Difference ({max_difference})')
plt.axhline(y=min_difference, color='g', linestyle='-', label=f'Min Difference ({min_difference})')
plt.xlabel('Index')
plt.ylabel('Percentage Difference')
plt.title('Percentage Difference over Time')
plt.legend()
plt.show()




# Load the data
df = pd.read_csv("aceusdt_price_comparison (1).csv")

# Find the max and min in the 'Percentage Difference' column
max_difference = df['Percentage Difference'].max()
min_difference = df['Percentage Difference'].min()

# Print max and min differences
print(f"Max Percentage Difference: {max_difference}")
print(f"Min Percentage Difference: {min_difference}")

# Plot the 'Percentage Difference'
plt.figure(figsize=(12, 6))
plt.plot(df['Percentage Difference'], label='Percentage Difference')
plt.axhline(y=max_difference, color='r', linestyle='-', label=f'Max Difference ({max_difference})')
plt.axhline(y=min_difference, color='g', linestyle='-', label=f'Min Difference ({min_difference})')
plt.xlabel('Index')
plt.ylabel('Percentage Difference')
plt.title('Percentage Difference over Time')
plt.legend()
plt.show()


```
